## Slicing lists

Values can be sliced from lists by specifying its position

When extracting a range of items, use [Start:End:Step]

```python
months = ['January', 'February', 'March', 'April', 'May', 'June']

# Try extracting January
print(months[0])

# Try extracting June
print(months[5])
```

```python
# Try extracting February to April
print(months[1:4])
```

```python
# Try extracting January to March
print(months[:3])
```

```python
# Try extracting every other item in the list
print(months[::2])
```

```python
# Try reversing the items in the list

print(months[5:1:-1])

empty = list()
print(empty)
```

## Dictionary

Dictionary is similar to a list, but the order of items doesn’t matter, and they aren’t selected by an offset such as 0 or 1. Instead, you specify a unique key to associate with each value. 

dictionary = { key : item }

```python
# Make your own dictionary here

my_dictionary_of_time = {
    'second' : 1,
    'minute' : 60,
    'hour' : 3600,
    'day' : 'a very large number'
}

print(my_dictionary_of_time)
```

```python
# Try accessing a value here

three_minutes = my_dictionary_of_time['minute'] * 3

print('Three minutes have', three_minutes, 'seconds.')
```

## For loops

Python makes frequent use of iterators, for good reason. They make it possible for you to traverse data structures without knowing how large they are or how they are implemented. 

It is possible to loop through:
1. Strings
2. Integers 
3. Lists
4. Etc

```python
# Try looping through a string
truth = "BioHackathon is awesome"

for alphabeth in truth:
    print(alphabeth)
```

```python
# Try looping through numbers

number = 10
for i in range(number):
    squared = i * i
    print(squared)
    
```

```python
# Try looping through lists

bacteria = ['Streptococcus', 'E. coli', 'Green sulphur bacteria']
for bact in bacteria:
    print(bact)

prime_numbers = [2, 3, 5, 7, 11]
for number in prime_numbers:
    multiply = number * 2
    print(multiply)
    
```

```python
number = 1000
for i in range(number):
    new_number = i * 2
    print(i, 'multiplied by two is', new_number)
```

DNA contains the genetic information of all living things. During cell division, DNA replication occurs. 

Given a strand of DNA, how do we find its complementary strand? 

Test case 1: ACGTACGTGACG

Answer: CGTCACGTACGT


Test case 2: ATCG

Answer: CGAT


Test case 3: GCGCGCGCGCGAAATTTT

Answer: AAAATTTCGCGCGCGCGC

```python
# Write your code here!

dna = 'ACGTACGTGACG'
bases_dictionary = {
    'A': 'T',
    'T': 'A',
    'G': 'C',
    'C': 'G'
}

complement_dna = ''

for base in dna:
    complement_dna += bases_dictionary[base]
reverse_complement = complement_dna[::-1]
print(reverse_complement)
```
